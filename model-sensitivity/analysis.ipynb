```python
import matplotlib.pyplot as plt

# Example data
data = {
    "GPT-4o": {"Review": 2.5, "Testing": 1.4 + 0.6 + 0.3},
    "GPT-3.5 Turbo": {"Review": 4.26, "Testing": 2.3},
    "Codellama-7B": {"Review": 12.5+1.3, "Testing": 0.7+1.7+2.3+2.3},
    "Mistral-7B-v0.1": {"Review": 9.1+0.7+0.3+2.7, "Testing": 2.5+0.8+2.3+1.4},
}

# Extract models and values for quant1 and quant2
models = list(data.keys())
quant1_values = [data[model]["Review"] for model in models]
quant2_values = [data[model]["Testing"] for model in models]

# Plotting the bar graph
x = range(len(models))  # X positions for the models

# Plot figure
plt.rc("font", family="serif", size=18, weight="bold")
plt.rc("axes", labelweight="bold", titleweight="bold")
plt.rc("text", usetex=True)
figure = plt.figure(figsize=(6,6), dpi=100)

bar_width = 0.2  # Set the bar width for both plots

plt.bar([i - bar_width/2 for i in x], quant1_values, width=bar_width, label="Review Time")
plt.bar([i + bar_width/2 for i in x], quant2_values, width=bar_width, label="Testing Time")


# Adding labels and title
#plt.xlabel("Models", fontsize=20)
plt.xticks(x, models, rotation=20)
plt.ylabel("Time (m)")
#plt.title("Developer Effort for Translating a Single File")
plt.xticks(x, models)
plt.legend()
plt.tight_layout()
plt.savefig("/Users/Akash/Desktop/model-sensitivity.png")
# Display the plot
plt.show()
```
