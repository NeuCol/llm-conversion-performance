```python
import matplotlib.pyplot as plt

# Example data
data = {
    "GPT-4o": {"Review": 10, "Testing": 20},
    "GPT-3.5 Turbo": {"Review": 100, "Testing": 200},
    "Codellama-7b": {"Review": 1000, "Testing": 2000},
    "Mistral-7b-v0.1": {"Review": 3000, "Testing": 5000},
}

# Extract models and values for quant1 and quant2
models = list(data.keys())
quant1_values = [data[model]["Review"] for model in models]
quant2_values = [data[model]["Testing"] for model in models]

# Plotting the bar graph
x = range(len(models))  # X positions for the models

# Plot figure
plt.rc("font", family="serif", size=16, weight="bold")
plt.rc("axes", labelweight="bold", titleweight="bold")
plt.rc("text", usetex=True)
figure = plt.figure(figsize=(8,6), dpi=100)

bar_width = 0.2  # Set the bar width for both plots

plt.bar([i - bar_width/2 for i in x], quant1_values, width=bar_width, label="Review")
plt.bar([i + bar_width/2 for i in x], quant2_values, width=bar_width, label="Testing")


# Adding labels and title
plt.xlabel("Models", fontsize=20)
plt.ylabel("Time", fontsize=20)
#plt.title("Developer Effort for Translating a Single File")
plt.xticks(x, models)
plt.legend()
plt.tight_layout()
plt.savefig("/Users/Akash/Desktop/model-sensitivity.png")
# Display the plot
plt.show()
```
